```python
words_in_sentences_set = open("Data/nouns_wiki_corpus.txt").read().splitlines()
# 3660755 / 35 = 104593
SET = int(3660755 // 5)
T = 1

words_in_sentences = words_in_sentences_set
```

```python
vocab = []
non_empty_words_in_sentences = []
for words_in_sentence in words_in_sentences:
    if(len(words_in_sentence) == 0): continue
    words = []
    for word in words_in_sentence.split():
        words.append(word)
        vocab.append(word)
    non_empty_words_in_sentences.append(words)

words_in_sentences = non_empty_words_in_sentences
vocab = set(vocab)
```

```python
import torch
import numpy as np
import random
from model import SentenceSimilarityModel
```

```python
EMBEDDING_DIM = 64
HIDDEN_DIM = 32

model_sentence_similarity = SentenceSimilarityModel(vocab, EMBEDDING_DIM, HIDDEN_DIM)

targets = []
for i in range(1, len(words_in_sentences)):
    targets.append(np.random.choice(words_in_sentences[i]))
targets.append(np.random.choice(words_in_sentences[0]))

s = []
t = []
random_data = np.random.randint(low=0, high=len(words_in_sentences), size = SET // 100)
random.shuffle(random_data)
for i in random_data:
  s.append(words_in_sentences[i])
  t.append(targets[i])

words_in_sentences = s
targets = t
```

```python
BATCH_SIZE = 32
EPOCHS = 5

model_sentence_similarity.train(words_in_sentences, targets, BATCH_SIZE, EPOCHS)
```

```python
torch.save(model_sentence_similarity, f"Model/model_sentence_similarity.pt")
```
