```python
import math
import os
import random

def min_index(lst):
    return min(range(len(lst)), key=lst.__getitem__)

def get_distance(v1, v2):
    return math.sqrt(sum([(a - b) ** 2 for a, b in zip(v1, v2)]))

def get_centroid(group, vector_list):
    centroids = []
    for i in range(len(group)):
        vectors = []
        for idx in group[i]:
            vectors.append(vector_list[idx][:-1])
        centroid = [0]*len(vectors[0])
        for vector in vectors:
            for j in range(len(vector)):
                centroid[j] += vector[j]
        for j in range(len(centroid)):
            centroid[j] /= len(vectors)
        centroids.append(centroid)
    return centroids
```

```python
folder_list = os.listdir('text')
print(folder_list)
text_files = []

for folder in folder_list:
    if(folder == '.DS_Store'):
        continue
    file_list = os.listdir('text/'+folder)
    text_files += [ ('text/'+folder+'/'+x) for x in file_list if x.endswith('.freq.txt')]
    break

cnt = 0
k = 3
MAX_ITER = 3
centroid_path = random.sample(text_files, k);

print(centroid_path)

centroid_vector = []
real_group = [[] for i in range(k)]
for centroid in centroid_path:
    vector = [y for y in open(centroid).read().split(',')]
    vector.pop(-1)
    vector = [int(y) for y in vector]
    vector.append(centroid)
    centroid_vector.append(vector)
```

```python
vector_list = []
for vector_path in text_files:
    vector = [y for y in open(vector_path).read().split(',') if y != '']
    vector.pop(-1)
    vector = [int(y) for y in vector if y != '']
    vector.append(vector_path)
    vector_list.append(vector)
```

```python
for iter in range(MAX_ITER):
  group = [[] for i in range(k)]
  idx1 = 0
  for vector_path in vector_list:
    idx1 += 1
    vector = vector_path
    li = []
    idx2 = 0
    for centroid in centroid_vector:
        idx2+= 1
        li.append(get_distance(vector[:-1], centroid))
        print('distance between centoid ' + str(idx1) + " " + str(idx2) + " =" + str(li[-1]))
    group[min_index(li)].append(idx1 - 1)
    print()
  centroid_vector = get_centroid(group, vector_list)
  real_group = group
  print(str((iter/k) * 100) + '%')
```

```python
group_path = ['group' + str(i + 1) + '.txt' for i in range(k)]

for i in range(len(real_group)):
  f = open(group_path[i], 'w')
  for path in group[i]:
    f.write(text_files[path] + '\n')
  f.close()
```
