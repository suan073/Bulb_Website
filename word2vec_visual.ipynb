```python
import numpy as np
from math import *
import numpy as np
import os

corpus_file = open('dataset.corpus.txt', 'w')
folder_list = os.listdir('text')
print(folder_list)
text_files = []

for folder in folder_list:
    if(folder == '.DS_Store'):
        continue
    file_list = os.listdir('text/'+folder)
    text_files += [ ('text/'+folder+'/'+x) for x in file_list if x.endswith('.txt')]

for file in text_files:
  readfile = open(file, 'r')
  while True:
    line = readfile.readline()
    if not line : break
    corpus_file.write(line)
  readfile.close()

corpus_file.close()
```

```python
import numpy as np
import pickle
import os
from util import create_contexts_target, to_cpu
import ptb

corpus, word_to_id, id_to_word = ptb.load_data('corpus')
freq_list = []
for file in text_files:
  vocablist = [0 for i in range(len(word_to_id))]
  words = open(file).read().replace('\n', '<eos>').strip().split()
  print(file)
  savepath = file[:-4] + '.freq.txt'
  #if os.path.exists(savepath):
  #  freq_list.append(savepath)
  #  continue
  for i, word in enumerate(words):
      if word in word_to_id:
          vocablist[word_to_id[word]] += 1
      elif word not in word_to_id:
          continue
  print(savepath)
  curfile = open(savepath, 'w')
  for i in vocablist:
      curfile.write(str(i)+',')
  curfile.close()
  
```

```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt

for file in text_files:
  open_path = file[:-4] + '.freq.txt'
  save_path = file[:-4] + '.png'
  plt.gca().set_yticklabels([])
  x_axis = np.array([i for i in range(len(word_to_id))])
  y_axis = [y for y in open(open_path).read().split(',')]
  y_axis.pop(-1)
  y_axis = np.array(y_axis)

  print(len(x_axis))
  print(len(y_axis))

  plt.plot(x_axis, y_axis)
  x = np.linspace(0, 5, 100)
  plt.title = (save_path[len(save_path) - 14:-4])
  plt.xlabel('Word ID')
  plt.ylabel('Frequency')
  plt.legend()
  plt.savefig(save_path, facecolor= '#eeeeee')
  plt.show()
  
```
